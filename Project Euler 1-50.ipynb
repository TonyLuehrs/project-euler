### General Use Functions

```python
def isPrime(n):
    if  n < 2:
        return False
    for i in range(2, int(n**.5) + 1):
        if n % i == 0:
            return False
    return True

def prod(l):
    p = 1
    for val in l:
        p *= int(val)
    return p

def divisors(n):
    divs = [i for i in range(1, int(n**.5) + 1) if n % i == 0]
    divs = divs + [n // d for d in divs]
    return list(set(divs))

def isPalindrome(s):
    return s == s[::-1]
```

# 9 - Largest Product in a Series
A Pythagorean triplet is a set of three natural numbers, a < b < c, for which,
a2 + b2 = c2

For example, 32 + 42 = 9 + 16 = 25 = 52.

There exists exactly one Pythagorean triplet for which a + b + c = 1000.
Find the product abc.


```python
def findPythSum(s):
    for a in range(1, s):
        for b in range(1, s):
            for c in range(1, s):
                if a + b + c == s:
                    if a**2 + b**2 == c**2:
                        return a*b*c

findPythSum(1000)                    
```

# 14 - Longest Collatz Sequence
The following iterative sequence is defined for the set of positive integers:

n → n/2 (n is even)
n → 3n + 1 (n is odd)

Using the rule above and starting with 13, we generate the following sequence:
13 → 40 → 20 → 10 → 5 → 16 → 8 → 4 → 2 → 1

It can be seen that this sequence (starting at 13 and finishing at 1) contains 10 terms. Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at 1.

Which starting number, under one million, produces the longest chain?

NOTE: Once the chain starts the terms are allowed to go above one million.


```python
collatzDict = {}

def collatz(n):
    steps = 1
    while n > 1:
        if n % 2 == 0:
            n //= 2
        else:
            n = 3*n + 1
        steps += 1
    return steps

collatz(13)
```

```python
collatzDict = {i:collatz(i) for i in range(1, 1000000)}
max(collatzDict, key=collatzDict.get)
```

```python
collatz(837799)
```

# 15 - Lattice Paths
Starting in the top left corner of a 2×2 grid, and only being able to move to the right and down, there are exactly 6 routes to the bottom right corner.

How many such routes are there through a 20×20 grid?


**In every $n\,x\,n$ grid, there are $2n$ steps to take to get to the opposite corner, and you need to choose which $n$ steps out of the $2n$ are the horizontal/vertical steps**

```python
from math import factorial as fac

def nCr(n, r):
    return fac(n) // (fac(r)*fac(n-r))

n = 80
nCr(2*n, n)
```

# 21 - Amicable Numbers
Let d(n) be defined as the sum of proper divisors of n (numbers less than n which divide evenly into n).
If d(a) = b and d(b) = a, where a ≠ b, then a and b are an amicable pair and each of a and b are called amicable numbers.

For example, the proper divisors of 220 are 1, 2, 4, 5, 10, 11, 20, 22, 44, 55 and 110; therefore d(220) = 284. The proper divisors of 284 are 1, 2, 4, 71 and 142; so d(284) = 220.

Evaluate the sum of all the amicable numbers under 10000.


```python
amicables = []

for i in range(10000):
    b = sum(divisors(i)) - i
    if sum(divisors(b)) - b == i and i != b:
        amicables.append(i)
print(amicables)
sum(amicables)
```

23 - Non-Abundant Sums
A perfect number is a number for which the sum of its proper divisors is exactly equal to the number. For example, the sum of the proper divisors of 28 would be 1 + 2 + 4 + 7 + 14 = 28, which means that 28 is a perfect number.

A number n is called deficient if the sum of its proper divisors is less than n and it is called abundant if this sum exceeds n.

As 12 is the smallest abundant number, 1 + 2 + 3 + 4 + 6 = 16, the smallest number that can be written as the sum of two abundant numbers is 24. By mathematical analysis, it can be shown that all integers greater than 28123 can be written as the sum of two abundant numbers. However, this upper limit cannot be reduced any further by analysis even though it is known that the greatest number that cannot be expressed as the sum of two abundant numbers is less than this limit.

Find the sum of all the positive integers which cannot be written as the sum of two abundant numbers.


```python
import numpy as np
abundants = np.array([i for i in range(1, 28124) if sum(divisors(i)) - i > i])
len(abundants)
```

**worst one in first 30, has a runtime of ~1 minute**

```python
allGood = []
failures = list(range(1, 28124))
for i, ab in enumerate(abundants):
    good = abundants[i:] + ab
    good = [g for g in good if g < 28124]
    allGood += good

allGood = list(set(allGood))
sum([f for f in failures if f not in allGood])
```

# 26 - Reciprocal Cycles


A unit fraction contains 1 in the numerator. The decimal representation of the unit fractions with denominators 2 to 10 are given:

    1/2	= 	0.5
    1/3	= 	0.(3)
    1/4	= 	0.25
    1/5	= 	0.2
    1/6	= 	0.1(6)
    1/7	= 	0.(142857)
    1/8	= 	0.125
    1/9	= 	0.(1)
    1/10	= 	0.1 

Where 0.1(6) means 0.166666..., and has a 1-digit recurring cycle. It can be seen that 1/7 has a 6-digit recurring cycle.

Find the value of d < 1000 for which 1/d contains the longest recurring cycle in its decimal fraction part.


```python
from decimal import getcontext
from decimal import Decimal as D

precision = 4502
third = (precision - 2) // 3
getcontext().prec = precision
```

```python
possDict = {}
for d in range(3, 1000):
    seq = D(1) / D(d)
    rs = str(seq)[::-1]
    #rint(rs, len(rs) - 2)
    if len(rs) > precision:
        rs = rs[1:4501]
        possDict[rs] = d
        #findCycleLength(rs)
```

```python
# remove 1 cycles
poss = list(possDict.keys())
for i, p in enumerate(poss):
    if all([dig == p[0] for dig in p[:third]]):
        poss.remove(p)

# remove 2 cycles
for i, p in enumerate(poss):
    if all([p[i:i+2] == p[0:2] for i in range(0, third, 2)]):
        poss.remove(p)
        
# remove n cycles
for n in range(3, third):
    for i, p in enumerate(poss):
        if all([p[i:i+n] == p[0:n] for i in range(0, third, n)]):
            poss.remove(p)
            best = p
possDict[best]
```

# 27 - Quadratic Primes
Euler discovered the remarkable quadratic formula:

$n^2+n+41$

It turns out that the formula will produce 40 primes for the consecutive integer values 0≤n≤39
. However, when $n=40$,$40^2+40+41=40(40+1)+41$ is divisible by 41, and certainly when $n=41$, $41^2+41+41$ is clearly divisible by 41.

The incredible formula $n^2−79n+1601$ was discovered, which produces 80 primes for the consecutive values 0≤n≤79

. The product of the coefficients, −79 and 1601, is −126479.

Considering quadratics of the form:

    $n^2+an+b$

, where |a|<1000 and |b|≤1000

where |n|
is the modulus/absolute value of n
e.g. |11|=11 and |−4|=4

Find the product of the coefficients, a
and b, for the quadratic expression that produces the maximum number of primes for consecutive values of n, starting with n=0.

```python
def f(n, a, b):
    return n**2 + a*n + b

best = 0
bestPair = (0, 0)
for a in range(-999, 1000):
    for b in range(-999, 1000, 2):
        n = 0
        while isPrime(f(n, a, b)):
            n += 1
        if n > best:
            best = n
            bestPair = (a, b)
print(best, bestPair, prod(bestPair))
```

# 30 - Digit Fifth Powers


Surprisingly there are only three numbers that can be written as the sum of fourth powers of their digits:

$1634 = 1^4 + 6^4 + 3^4 + 4^4$  
$8208 = 8^4 + 2^4 + 0^4 + 8^4$  
$9474 = 9^4 + 4^4 + 7^4 + 4^4$  

As $1 = 1^4$ is not a sum it is not included.

The sum of these numbers is 1634 + 8208 + 9474 = 19316.

Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.


```python
hits = []
for i in range(2, 1000000):
    if sum([int(dig)**5 for dig in str(i)]) == i:
        hits.append(i)
sum(hits)
```

In England the currency is made up of pound, £, and pence, p, and there are eight coins in general circulation:

    1p, 2p, 5p, 10p, 20p, 50p, £1 (100p) and £2 (200p).

It is possible to make £2 in the following way:

    1×£1 + 1×50p + 2×20p + 1×5p + 1×2p + 3×1p

How many different ways can £2 be made using any number of coins?


```python
ways = 1 # count the way of having a 2 dollar coin
for a in range(2+1):
    for b in range(4+1):
        for c in range(10+1):
            for d in range(20+1):
                for e in range(40+1):
                    for f in range(100+1):
                        if a*100 + b*50 + c*20 + d*10 + e*5 + f*2 <= 200:
                            ways += 1
ways
```

# Pandigital Products


We shall say that an n-digit number is pandigital if it makes use of all the digits 1 to n exactly once; for example, the 5-digit number, 15234, is 1 through 5 pandigital.

The product 7254 is unusual, as the identity, 39 × 186 = 7254, containing multiplicand, multiplier, and product is 1 through 9 pandigital.

Find the sum of all products whose multiplicand/multiplier/product identity can be written as a 1 through 9 pandigital.
HINT: Some products can be obtained in more than one way so be sure to only include it once in your sum.


```python
def isPandigital(n):
    divPairs = []
    for i in range(2, int(n**.5)+1):
        if n % i == 0:
            divPairs.append((i, n // i))
    digsUsed = [''.join(sorted(str(n) + str(p[0]) + str(p[1]))) for p in divPairs]
    if '123456789' in digsUsed:
        return True
    

# threw it in a while loop first, doesn't look like there are any after 10000
n = 1000
pans = []
while n < 10000:
    if isPandigital(n):
        pans.append(n)
    n += 1
    
sum(pans)
```

# 33 - Digit Cancelling Fractions


The fraction 49/98 is a curious fraction, as an inexperienced mathematician in attempting to simplify it may incorrectly believe that 49/98 = 4/8, which is correct, is obtained by cancelling the 9s.

We shall consider fractions like, 30/50 = 3/5, to be trivial examples.

There are exactly four non-trivial examples of this type of fraction, less than one in value, and containing two digits in the numerator and denominator.

If the product of these four fractions is given in its lowest common terms, find the value of the denominator.


```python
successes = []
for n in range(10, 100):
    for d in range(10, 100):
        if n != d:
            sn, sd = str(n), str(d)

            if sn[0] == sd[0]:
                if sd[1] != '0' and int(sn[1]) / int(sd[1]) == n/d:
                    successes.append((n, d))

            if sn[0] == sd[1]:
                if int(sn[1]) / int(sd[0]) == n/d:
                    successes.append((n, d))

            if sn[1] == sd[0]:
                if sd[1] != '0' and int(sn[0]) / int(sd[1]) == n/d:
                    successes.append((n, d))

            if sn[1] == sd[1] and sn[1] != '0':
                if int(sn[0]) / int(sd[0]) == n/d:
                    successes.append((n, d))
```

```python
from fractions import Fraction

successes = sorted([sorted(s) for s in successes])[::2]

print(successes)
num = prod([s[0] for s in successes])
denom = prod([s[1] for s in successes])

print(num, denom)
Fraction(num, denom)
```

# 35 - Circular Primes


The number, 197, is called a circular prime because all rotations of the digits: 197, 971, and 719, are themselves prime.

There are thirteen such primes below 100: 2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79, and 97.

How many circular primes are there below one million?


```python
def getRotations(n):
    sn = str(n)
    return [int(sn[-i:] + sn[:-i]) for i in range(len(sn))]

def isCircularPrime(n):
    sn = str(n)
    return all([isPrime(int(sn[-i:] + sn[:-i])) for i in range(len(sn))])
    
getRotations(197)
```

# 37 - Truncatable Primes


The number 3797 has an interesting property. Being prime itself, it is possible to continuously remove digits from left to right, and remain prime at each stage: 3797, 797, 97, and 7. Similarly we can work from right to left: 3797, 379, 37, and 3.

Find the sum of the only eleven primes that are both truncatable from left to right and right to left.

NOTE: 2, 3, 5, and 7 are not considered to be truncatable primes.


```python
def getTruncations(n):
    fromLeft = [int(str(n)[i:]) for i in range(1, len(str(n)))]
    fromRight = [int(str(n)[:i]) for i in range(1, len(str(n)))]
    return fromLeft + fromRight
    
getTruncations(3797)
```

What is the largest 1 to 9 pandigital 9-digit number that can be formed as the concatenated product of an integer with (1,2, ... , n) where n > 1?


```python
def isPanDig(sn):
    return ''.join(sorted(sn)) == '123456789'

for n in range(2, 10000000):
    allDigits = str(n)
    mult = 2
    while len(allDigits) < 9:
        allDigits += str(n*mult)
        mult += 1
    if len(allDigits) == 9:
        if isPanDig(allDigits):
            print(n, allDigits)
```

# 43 - Sub-String Divisibility


The number, 1406357289, is a 0 to 9 pandigital number because it is made up of each of the digits 0 to 9 in some order, but it also has a rather interesting sub-string divisibility property.

Let d1 be the 1st digit, d2 be the 2nd digit, and so on. In this way, we note the following:

    d2d3d4=406 is divisible by 2
    d3d4d5=063 is divisible by 3
    d4d5d6=635 is divisible by 5
    d5d6d7=357 is divisible by 7
    d6d7d8=572 is divisible by 11
    d7d8d9=728 is divisible by 13
    d8d9d10=289 is divisible by 17

Find the sum of all 0 to 9 pandigital numbers with this property.


```python
from itertools import permutations
digs = '0123456789'
panDigs = [''.join(p) for p in permutations(digs, len(digs)) if p[0] != '0']

total = 0
primes = [2, 3, 5, 7, 11, 13, 17]
for pd in panDigs:
    if all([int(pd[1+i: 4+i]) % primes[i] == 0 for i in range(len(primes))]):
        total += int(pd)
        print(int(pd))
total
```

Pentagonal numbers are generated by the formula, Pn=n(3n−1)/2. The first ten pentagonal numbers are:

1, 5, 12, 22, 35, 51, 70, 92, 117, 145, ...

It can be seen that P4 + P7 = 22 + 70 = 92 = P8. However, their difference, 70 − 22 = 48, is not pentagonal.

Find the pair of pentagonal numbers, Pj and Pk, for which their sum and difference are pentagonal and D = |Pk − Pj| is minimised; what is the value of D?

**Pretty bad, took about 5 minutes**


```python
N = list(range(1,4000))
Pn = [n*(3*n-1) // 2 for n in N]
```

```python
from itertools import combinations
from math import fabs

pairs = list(combinations(Pn, 2))
add = [p[0] + p[1] for p in pairs]
sub = [fabs(p[0] - p[1]) for p in pairs]

both = []
for i, pair in enumerate(pairs):
    if add[i] in Pn and sub[i] in Pn:
        both.append(pair)
diffs = [fabs(p[0] - p[1]) for p in both]
min(diffs)
```

# 45 - Triangular, Pentagonal, Hexagonal


Triangle, pentagonal, and hexagonal numbers are generated by the following formulae:  
Triangle 	  	Tn=n(n+1)/2 	  	1, 3, 6, 10, 15, ...  
Pentagonal 	  	Pn=n(3n−1)/2 	  	1, 5, 12, 22, 35, ...  
Hexagonal 	  	Hn=n(2n−1) 	  	1, 6, 15, 28, 45, ...  

It can be verified that T285 = P165 = H143 = 40755.

Find the next triangle number that is also pentagonal and hexagonal.


```python
def triGen():
    def t_n(n): return n*(n+1)//2
    n = 1
    while True:
        yield t_n(n)
        n += 1

def pentGen():
    def p_n(n): return n*(3*n-1)//2
    n = 1
    while True:
        yield p_n(n)
        n += 1

def hexGen():
    def h_n(n): return n*(2*n-1)
    n = 1
    while True:
        yield h_n(n)
        n += 1
    
```

```python
def findMatchStartingAt(tIter, pIter, hIter):
    
    # initialize generators
    tG = triGen()
    pG = pentGen()
    hG = hexGen()
    
    # iterate through initial iterations
    for i in range(tIter):
        t = next(tG)
    for i in range(pIter):
        p = next(pG)
    for i in range(hIter):
        h = next(hG)
    
    # find next match
    while True:
        while h >= p and h >= t:
            while p > t:
                t = next(tG)
                if h == p == t: return h
            if h > p and t >= p:
                p = next(pG)
                if h == p == t: return h
        h = next(hG)
        if h == p == t: return h
        
```

# Goldbach's Other Conjecture


It was proposed by Christian Goldbach that every odd composite number can be written as the sum of a prime and twice a square.

$9 = 7 + 2×1^2$  
$15 = 7 + 2×2^2$  
$21 = 3 + 2×3^2$  
$27 = 19 + 2×2^2$  
$33 = 31 + 2×1^2$  

It turns out that the conjecture was false.

What is the smallest odd composite that cannot be written as the sum of a prime and twice a square?


```python
def isOddComp(n):
    return n%2 == 1 and not isPrime(n)

def f(p, b):
    return p + 2*(b**2)
```

# 47 - Distinct Prime Factors


The first two consecutive numbers to have two distinct prime factors are:

14 = 2 × 7
15 = 3 × 5

The first three consecutive numbers to have three distinct prime factors are:

644 = 2² × 7 × 23
645 = 3 × 5 × 43
646 = 2 × 17 × 19.

Find the first four consecutive integers to have four distinct prime factors each. What is the first of these numbers?



```python
def getFacsPows(n):
    fpD = {}
    while n != 1:
        i = 2
        while True:
            if n % i == 0:
                if i not in fpD:
                    fpD[i] = 1
                else:
                    fpD[i] += 1
                n //= i
                break
            else:
                i += 1
    return fpD
[(k, v) for k, v in getFacsPows(15).items()]
```

```python
match = 4
n = 648

while True:
    skip = 1
    a = [k**v for k, v in getFacsPows(n).items()]
    if len(a) == 4:
        skip += 1
        b = [k**v for k, v in getFacsPows(n+1).items()]
        if len(b) == 4:
            skip += 1
            c = [k**v for k, v in getFacsPows(n+2).items()]
            if len(c) == 4:
                skip += 1
                d = [k**v for k, v in getFacsPows(n+3).items()]
                if len(list(set(a + b + c + d))) == match**2:
                    print(n, a, b, c, d)
                    break
    n += skip
```
